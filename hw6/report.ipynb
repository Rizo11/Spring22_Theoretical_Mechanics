# Task 1

## Description

There are 3 weights with masses $M_1 = 20$ kg, $M_2 = 15$ kg, $M_3 = 10$ they are connected by ideal string. This string go through by two pulleys L, N. When the $M_1$ weight goes down on 1 meter, the body ABCD shifts on some distance $s$. The task is to find the distance of this movement according to the ground. Neglect the friction between the floor and ABCD.

## Solution

### R.O

1 Body: $M_1$, translatory motion

2 Body: $M_2$, translatory motion

3 Body: $M_3$, translatory motion

4 Body: $ABCD$, translatory motion

Method for center of masses will be used.

### Conditions

Along $x$-axis, $d = 1$
$$
\begin{bmatrix}
body & initial & final\\
x_{ABCD} & x_0 & x_0 + s\\
x_{m_1} & x_1 & x_1 + s\\
x_{m_2} & x_2 & x_3 + s + d\\
x_{m_3} & x_3 & x_3 + s + d cos{(\pi/3)}\\
\dot x_c & 0 & 0\\
\ddot x_c & 0 & 0\\
\end{bmatrix}
$$

### Force analysis

As show in the picture we have 5 forces:
1. gravitation force of $M_1$, $G_1$
1. gravitation force of $M_2$, $G_2$
1. gravitation force of $M_3$, $G_3$
1. gravitation force of trapezoidal construction  $M_{ABCD}$, $G_{ABCD}$
2. normal force of the system $N$

### Solution

Center of mass of the system did not move, and no acceleration. So we get

$$
\begin{equation}
M \ddot x_{c} = 0
\end{equation}
$$

$$
\begin{equation}
M \dot x_{c} = 0
\end{equation}
$$

$$
\begin{equation}
M x_{c} = 0
\end{equation}
$$

As center of mass did not move, final and initial center of masses of the system are equal:

$$
\begin{equation}
M x_{c_i} = M x_{c_f}
\end{equation}
$$

Writing expanding equiton of each part of the system, we get:

$$
\sum {m_j x_{j_i}} = \sum {m_j x_{j_f}}
$$

$$
\begin{equation}
M_1 s + M_2 d + M_2 s + M_3 d sin(\pi/6) + M_3 s + M_{ABCD}s = 0
\end{equation}
$$

where $d = 1$.

Solve for $s$, and get:

$$
\begin{equation}
s = -\frac{M_2 d + M_3 d sin(\pi/6)}{M_1  + M_2 + M_3 + M_{ABCD}}
\end{equation}
$$

As the weight of the $ACBD$ is not give, let's assume that it is $M_{ABCD} = 97.85 kg$,

then $s = -14.0007 cm$

## Highlited Answers

$s = 14.00007$ $cm$ to the left

# Task 2

## Description

On the figure a pulley with a rope running over it is presented. A man (mass $m$) holds one side of the rope at $A$ while a load, equal to the weight of the man, is attached to the other end of the rope at $B$. What would happen to the load if the man starts climbing up the rope with the velocity $a$ relative to the rope. The weight of the pulley is equal $0.25$. The mass of the pulley is distributed uniformly along its rim. Answer: The load will ascend with velocity $\frac{4a}{9}$

## Solution

### R.O

1. pulley: rotation
2. box: rectilinear
3. man: rectilinear

### Conditions

$$
\begin{bmatrix}
object & initial & final\\
y_{A} & y_{A_0} & ?\\
y_{B} & y_{B_0} & ?\\
v_{A} & 0 & v + a\\
v_{B} & 0 & -v\\
a_{A} & -g & -g\\
a_{B} & -g & -g
\end{bmatrix}
$$

### Force analysis

As shown in the picture
1. $G_A$
2. $G_B$
3. $G_{p}$ - pulley gravity
4. $R$ reaction force of pulley

Theorem on change of angular momentum of a system.

$$
\begin{equation}
\frac{d\vec L_c}{dt} = \sum { \vec M_c}
\end{equation}
$$

### Solution

The pulley experiences a torque due to the tension in the rope, which causes it to rotate. Let $\tau$ represent this torque. According to Newton's second law for rotational motion, the torque $\tau$ is equal to the rate of change of angular momentum $\frac{dL}{dt}$, where $L$ is the angular momentum.

$\tau = \frac{dL}{dt}$

Men and load have the same mass, thus cause torques which are the same in magnitude, but different in directions.

$ -m_a g r + m_b g r = 0$, where $r$ is the radius of pulley.

$0 = L_{p} + L_A + L_B$

$0 = Iw + m_a \cdot v_a \cdot r + m_b \cdot v_b \cdot r$, where $m_b = m_a = m$, where $I$ is inertia of a ring-like shape.

$0 = \frac{mr^2}{4} \frac{v_b}{r} + m (v_b + a) \cdot r + m \cdot v_b \cdot r$

$v_b = - \frac{4a}{9}$, B will move upwards.

## Highlited answers

$v_b = - \frac{4a}{9}$, B will move upwards.
