# Task 1

## Description

There are 3 weights with masses $M_1 = 20$ kg, $M_2 = 15$ kg, $M_3 = 10$ they are connected by ideal string. This string go through by two pulleys L, N. When the $M_1$ weight goes down on 1 meter, the body ABCD shifts on some distance $s$. The task is to find the distance of this movement according to the ground. Neglect the friction between the floor and ABCD.

## Solution

### R.O

1 Body: $M_1$, translatory motion

2 Body: $M_2$, translatory motion

3 Body: $M_3$, translatory motion

4 Body: $ABCD$, translatory motion

Method for center of masses will be used.

### Conditions

Along $x$-axis, $d = 1$
$$
\begin{bmatrix}
body & initial & final\\
x_{ABCD} & x_0 & x_0 + s\\
x_{m_1} & x_1 & x_1 + s\\
x_{m_2} & x_2 & x_3 + s + d\\
x_{m_3} & x_3 & x_3 + s + d cos{(\pi/3)}\\
\dot x_c & 0 & 0\\
\ddot x_c & 0 & 0\\
\end{bmatrix}
$$

### Force analysis

As show in the picture we have 5 forces:
1. gravitation force of $M_1$, $G_1$
1. gravitation force of $M_2$, $G_2$
1. gravitation force of $M_3$, $G_3$
1. gravitation force of trapezoidal construction  $M_{ABCD}$, $G_{ABCD}$
2. normal force of the system $N$

### Solution

Center of mass of the system did not move, and no acceleration. So we get

$$
\begin{equation}
M \ddot x_{c} = 0
\end{equation}
$$

$$
\begin{equation}
M \dot x_{c} = 0
\end{equation}
$$

$$
\begin{equation}
M x_{c} = 0
\end{equation}
$$

As center of mass did not move, final and initial center of masses of the system are equal:

$$
\begin{equation}
M x_{c_i} = M x_{c_f}
\end{equation}
$$

Writing expanding equiton of each part of the system, we get:

$$
\sum {m_j x_{j_i}} = \sum {m_j x_{j_f}}
$$

$$
\begin{equation}
M_1 s + M_2 d + M_2 s + M_3 d sin(\pi/6) + M_3 s + M_{ABCD}s = 0
\end{equation}
$$

where $d = 1$.

Solve for $s$, and get:

$$
\begin{equation}
s = -\frac{M_2 d + M_3 d sin(\pi/6)}{M_1  + M_2 + M_3 + M_{ABCD}}
\end{equation}
$$

As the weight of the $ACBD$ is not give, let's assume that it is $M_{ABCD} = 97.85 kg$,

then $s = -14.0007 cm$

$s = 14.00007$ $cm$ to the left
